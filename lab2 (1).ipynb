```python
from collections import deque
print("STACK IMPLIMENTATION")
stack=deque()
stack.append('A')
stack.append('B')
stack.append('C')
print("Stack pushing A,B,C:",list(stack))
print("Poped from stack:",stack.pop())
print("Stack after popping:",list(stack))
print("\nQUEUE IMPLIMENTATION")
queue=deque()
```

Queue Implementation

```python
print("\nOUEUE IMPLIMENTATION")
queue=deque()
queue.append('1')
queue.append('2')
queue.append('3')
print("Queue afterenqueuing 1,2,3:",list(queue))
print("Dequeued from queue:",queue.popleft())
print("Queue after dequeuing:",list(queue))
```

```python
QUEUE IMPLIMENTATION
Queue afterenqueuing 1,2,3: ['1', '2', '3']
Dequeued from queue: 1
Queue after dequeuing: ['2', '3']
```

List to store the heap

```python
import heapq
priority_queue=[]
def add_student(name,score):
  heapq.heappush(priority_queue,(-score,name))
print("Added:{name}with score{score}")
```

Function to get the student woth hightest priority(highest score

```python
def pop_top_student():
  if priority_queue:
    score,name=heapq.heappop(priority_queue)
    print(f"Selected for top priority:{name}with score{-score}")
  else:
       print("No students in the queue.")
add_student("Alice",92)
add_student("Bob",85)
add_student("Charlie",95)
add_student("Diana",88)
print("\n")
pop_top_student()
pop_top_student()
pop_top_student()
pop_top_student()
```
